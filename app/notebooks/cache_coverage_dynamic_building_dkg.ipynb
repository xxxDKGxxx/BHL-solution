```python
import json

from app.testers.LLM_cache_building_tester import LlmCacheBuildingTester

with open('programming_questions.json', 'r') as f:
	data = json.load(f)

data
```

```python

questions = [question["original_question"] for question in data]
similair_questions = []

for question in data:
	for similair_question in question["paraphrases"]:
		similair_questions.append(similair_question)

print(questions, similair_questions)

similair_questions = random.choices(similair_questions, k=10)
```

```python
tester = LlmCacheBuildingTester(
	questions,
	similair_questions,
	name="programming_questions"
)

tester.run()
```
