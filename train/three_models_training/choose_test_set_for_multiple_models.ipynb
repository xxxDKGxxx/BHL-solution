```python
import os
import sys

project_root = os.path.abspath(os.path.join(os.getcwd(), '..', '..'))
app_root = os.path.abspath(os.path.join(project_root, '../../app', '..'))

if project_root not in sys.path:
	sys.path.append(project_root)
	sys.path.append(app_root)
from train.datasets_preprocessing.datasets_preprocessing import load_json_data, make_pipeline
import pandas as pd

math_pipeline = make_pipeline('math')

X_json_raw = load_json_data('../datasets_preprocessing/datasets/math')
math_pipeline.fit_transform(X_json_raw)

math_df = pd.read_csv(os.path.join('../datasets_preprocessing/csv_question_files', 'math.csv'))
math_df.drop_duplicates(subset=["question"], inplace=True)

bio_pipeline = make_pipeline('bio')

X_json_raw = load_json_data('../datasets_preprocessing/datasets/bio')
bio_pipeline.fit_transform(X_json_raw)
bio_df = pd.read_csv(os.path.join('../datasets_preprocessing/csv_question_files', 'bio.csv'))
bio_df.drop_duplicates(subset=["question"], inplace=True)

code_pipeline = make_pipeline('code')

X_json_raw = load_json_data('../datasets_preprocessing/datasets/code')
code_pipeline.fit_transform(X_json_raw)

code_df = pd.read_csv(os.path.join('../datasets_preprocessing/csv_question_files', 'code.csv'))
code_df.drop_duplicates(subset=["question"], inplace=True)
```

```python
import pandas as pd

bio_train_df = pd.read_csv("saved_models/bio/train_set.csv", index_col=0)
math_train_df = pd.read_csv("saved_models/math/train_set.csv", index_col=0)
code_train_df = pd.read_csv("saved_models/code/train_set.csv", index_col=0)
one_model_train_df = pd.read_csv("saved_models/1 model/train_set.csv", index_col=0)


all_train_dfs = (pd.concat([bio_train_df, math_train_df, code_train_df, one_model_train_df], axis=0).reset_index
                 (drop=True))
all_train_dfs.drop_duplicates(subset=["question"], keep="first", inplace=True)
all_train_dfs
```

```python
def filter_df_to_not_contain_values_that_other_df_does(df: pd.DataFrame, other_df: pd.DataFrame, column_name: str) ->\
		pd.DataFrame:
	return df[~df[column_name].isin(other_df[column_name])]


math_df_filtered = filter_df_to_not_contain_values_that_other_df_does(math_df, all_train_dfs, column_name="question")

math_df_filtered
```

```python
bio_df_filtered = filter_df_to_not_contain_values_that_other_df_does(bio_df, all_train_dfs, column_name="question")

bio_df_filtered
```

```python
code_df_filtered = filter_df_to_not_contain_values_that_other_df_does(code_df, all_train_dfs, column_name="question")

code_df_filtered
```

```python
for i in range(3):
	sample = pd.concat([math_df_filtered.sample(n=2000), bio_df_filtered.sample(n=2000), code_df_filtered.sample
	(n=2000)], axis=0)
	sample.drop_duplicates(subset=["question"], inplace=True)
	sample.to_csv("../datasets_preprocessing/test_all_models/test_" + str(i) + ".csv")
```
