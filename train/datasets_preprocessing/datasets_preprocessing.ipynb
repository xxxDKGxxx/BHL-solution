# Preprocessing danych używany w celu modyfikacji zbiorów danych z formacie .jsonl, przepisując je do formatu .csv

```python
import os
import sys

project_root = os.path.abspath(os.path.join(os.getcwd(), '../../train', '..'))

if project_root not in sys.path:
    sys.path.append(project_root)
```

## Generacja zbioru danych dla pytań matematycznych

```python
from train.datasets_preprocessing.datasets_preprocessing import load_json_data, make_pipeline
import pandas as pd

math_pipeline = make_pipeline('math')

X_json_raw = load_json_data('./datasets/math')
math_pipeline.fit_transform(X_json_raw)

df = pd.read_csv(os.path.join('csv_question_files', 'math.csv'))
df['question'].value_counts(ascending=False)
```

## Generacja zbioru danych dla pytań biologicznych

```python
bio_pipeline = make_pipeline('bio')

X_json_raw = load_json_data('./datasets/bio')
bio_pipeline.fit_transform(X_json_raw)

df = pd.read_csv(os.path.join('csv_question_files', 'bio.csv'))
df['question'].value_counts(ascending=False)
```

## Generacja zbioru danych dla pytań programistycznych

```python
code_pipeline = make_pipeline('code')

X_json_raw = load_json_data('./datasets/code')
code_pipeline.fit_transform(X_json_raw)

df = pd.read_csv(os.path.join('csv_question_files', 'code.csv'))
df['question'].value_counts(ascending=False)
```
