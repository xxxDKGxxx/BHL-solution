```python
import random
import sys
import os

# Pobieramy ścieżkę do katalogu głównego projektu (dwa poziomy wyżej: ../..)
project_root = os.path.abspath(os.path.join(os.getcwd(), '..', '..'))

# Dodajemy ją do ścieżek Pythona, jeśli jej tam nie ma
if project_root not in sys.path:
    sys.path.append(project_root)
```

```python
import json

from app.testers.LLM_cache_building_tester import LlmCacheBuildingTester

with open('programming_questions_eng.json', 'r') as f:
	prog_eng_data = json.load(f)

prog_eng_data
```

```python

prog_eng_questions = [question["original_question"] for question in prog_eng_data]
prog_eng_similair_questions = []

for question in prog_eng_data:
	for similair_question in question["paraphrases"]:
		prog_eng_similair_questions.append(similair_question)

prog_eng_similair_questions = random.choices(prog_eng_similair_questions, k=10)

print(prog_eng_questions, prog_eng_similair_questions)
```

```python
prog_eng_tester = LlmCacheBuildingTester(
	prog_eng_questions,
	prog_eng_similair_questions,
	name="programming_questions_eng"
)

prog_eng_tester.run()
```

```python
from app.testers.LLM_cache_building_tester import summarise_batch_reports

df = summarise_batch_reports(name="programming_questions_eng")

df.drop_duplicates(subset=["question"], inplace=True)

df["is_cached"].value_counts(normalize=True)
```
